## Aproximación de Euler

```python
def f(t, y):
    return y-t**2+1
a = 0
b = 2
N = 10

y0 = 0.5
```

```python
resultado = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=N)
resultado
```

## Comparación con la solución real

La solución real de la ecuación diferencial es: $y(t) = (t+ 2)^2 - 0.5 e^t$

1. Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.

```python
import matplotlib.pyplot as plt
import numpy as np

ys = resultado[0]
ts = resultado[1]

t_real = np.linspace(a, b, 100)
y_real = (t_real+1)**2 - 0.5*np.exp(t_real)
plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker='o', linestyle='-', label='Aproximación Euler', color='orchid')
plt.plot(t_real, y_real, linestyle='--', label='Solución Real', color='skyblue')
plt.title('Gráfica de y vs t usando el Método de Euler', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ys = resultado[0]
ts = resultado[1]
resultado2 = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=5)
ys2 = resultado2[0]
ts2 = resultado2[1]
resultado3 = ODE_euler(a=a, b=b,f=f, y_t0=y0 ,N=20)
ys3 = resultado3[0]
ts3 = resultado3[1]


t_real = np.linspace(a, b, 100)
y_real = (t_real+1)**2 - 0.5*np.exp(t_real)
plt.figure(figsize=(8, 5))
plt.plot(ts, ys, marker='o', linestyle='-', label='Aproximación Euler N = 10', color='lightcoral')
plt.plot(ts2, ys2, marker='o', linestyle='-', label='Aproximación Euler N = 5', color='olive')
plt.plot(ts3, ys3, marker='o', linestyle='-', label='Aproximación Euler N = 20', color='violet')
plt.plot(t_real, y_real, linestyle='--', label='Solución Real', color='skyblue')
plt.title('Gráfica de y vs t usando el Método de Euler', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)
plt.show()
```

3. Repetir 2do ejercicio con solución real:  $y_{real} = \frac{t}{1+ ln(t)}$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(t, y):
    return y/t - (y/t)**2

resultado = ODE_euler(a=1, b=b,f=f, y_t0=1 ,N=N)
ys = resultado[0]
ts = resultado[1]
resultado2 = ODE_euler(a=1, b=b,f=f, y_t0=1 ,N=5)
ys2 = resultado2[0]
ts2 = resultado2[1]
resultado3 = ODE_euler(a=1, b=b,f=f, y_t0=1 ,N=20)
ys3 = resultado3[0]
ts3 = resultado3[1]


t_real = np.linspace(1, b, 400)
y_real = (t_real)/(1+np.log(t_real))


plt.figure(figsize=(8, 6))
plt.plot(ts, ys, marker='o', linestyle='-', label='Aproximación Euler N = 10', color='lightcoral')
plt.plot(ts2, ys2, marker='o', linestyle='-', label='Aproximación Euler N = 5', color='olive')
plt.plot(ts3, ys3, marker='o', linestyle='-', label='Aproximación Euler N = 20', color='violet')
plt.plot(t_real, y_real, linestyle='--', label='Solución Real', color='skyblue')
plt.title('Gráfica de y vs t usando el Método de Euler', fontsize=14)
plt.xlabel('t', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=10)
plt.show()
```

4. ¿Qué pasa al aumentar N?

Al aumentar más el valor de N, el método de Euler se aproxima más a la solución real.
